```python
import requests
from bs4 import BeautifulSoup
from fpdf import FPDF

# Function to scrape text from a given URL
def scrape_text(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    # Extract text content from paragraphs and headers
    text = ''
    for tag in soup.find_all(['p', 'h1', 'h2', 'h3', 'h4', 'h5', 'h6']):
        text += tag.get_text() + '\n'
    return text

# Function to generate a PDF from the scraped text
def generate_pdf(content, file_name='output.pdf'):
    pdf = FPDF()
    pdf.set_auto_page_break(auto=True, margin=15)
    pdf.add_page()

    pdf.set_font("Arial", size=12)

    for line in content.split('\n'):
        pdf.cell(200, 10, txt=line.encode('latin-1', 'replace').decode('latin-1'), ln=True)

    pdf.output(file_name)
    print(f"PDF generated: {file_name}")

# Main function to crawl through subpages and gather content
def scrape_website(base_url):
    visited_urls = set()
    all_text = ''

    def scrape_page(url):
        nonlocal all_text
        if url not in visited_urls:
            print(f"Scraping: {url}")
            visited_urls.add(url)
            all_text += scrape_text(url)

            # Extracting links to other subpages
            response = requests.get(url)
            soup = BeautifulSoup(response.text, 'html.parser')
            for link in soup.find_all('a', href=True):
                href = link['href']
                if href.startswith('/'):
                    href = base_url + href
                if href.startswith(base_url):
                    scrape_page(href)

    scrape_page(base_url)
    return all_text

# Base URL for scraping
base_url = 'https://www.srh-hochschule-heidelberg.de/en/'

# Scrape the website
text_data = scrape_website(base_url)

# Generate PDF with the scraped text
generate_pdf(text_data)
```
